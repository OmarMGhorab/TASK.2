```python
first_model = cv2.imread('C:\\Users\\Dell\\Pictures\\elsesy.jpg')
first_model = cv2.cvtColor(first_model, cv2.COLOR_BGR2RGB)
first_gray =cv2.cvtColor(first_model, cv2.COLOR_RGB2GRAY)

norm_model=first_gray/255.0
def ft_first_model(norm_model):
    f = np.fft.fft2(norm_model)
    fshift = np.fft.fftshift(f)
    frequency_tx = 20*np.log(np.abs(fshift))
    
    return  frequency_tx
f_first_model = ft_first_model(norm_model)
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10)) 
ax1.imshow(first_model)
ax2.imshow(f_first_model, cmap='gray') 
```
